```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.columns = [col.strip() for col in df.columns]  

df.head()
```

```python
print("First 5 rows of the dataset:")
print(df.head())
```

```python
print("\nData types and missing values:")
print(df.info())
```

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
print("\nDescriptive statistics:")
print(df.describe())
```

```python
grouped_means = df.groupby('species').mean()
print("\nMean of each feature grouped by species:")
print(grouped_means)
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
for species in df['species'].unique():
    subset = df[df['species'] == species].reset_index()
    plt.plot(subset.index, subset['sepal length (cm)'], label=species)
plt.title('Sepal Length Trend (by sample index)')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
grouped_means['petal length (cm)'].plot(kind='bar', color='skyblue')
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['petal width (cm)'], bins=15, color='green', edgecolor='black')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("\nObservations:")
print("- Iris-setosa has the shortest petal and sepal lengths.")
print("- Iris-virginica generally has the largest features.")
print("- There’s a clear positive correlation between sepal length and petal length.")
```
